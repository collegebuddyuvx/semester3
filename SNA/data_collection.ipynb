```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Create a realistic social network dataset
df = pd.DataFrame(
    {
        "Source": [
            "Alice",
            "Bob",
            "Charlie",
            "David",
            "Eve",
            "Alice",
            "Charlie",
            "Eve",
        ],
        "Destination": [
            "Bob",
            "Charlie",
            "Alice",
            "Eve",
            "Alice",
            "David",
            "David",
            "Bob",
        ],
    }
)

# Build the network using NetworkX
G = nx.from_pandas_edgelist(
    df, source="Source", target="Destination", create_using=nx.DiGraph()
)

# Using a directed graph to represent one-way relationships (e.g., Alice follows Bob)
plt.figure(figsize=(8, 6))
nx.draw(
    G,
    with_labels=True,
    node_color="lightgreen",
    node_size=800,
    font_size=10,
    font_weight="bold",
    arrows=True,
)
plt.title("Social Network Connections")
plt.show()
```
